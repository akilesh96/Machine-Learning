```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
main=np.zeros((12,18,3))
main[4:8,:,:]=1
main[:4,:,1:2]=1
main[::,:3,0]=1
main[::,:3,1:]=0
plt.imshow(main)
```

```python
main=np.ones((12,18,3))
main[6::,::,1:]=0
X=np.array([[-0.5,-0.5], [-1,12], [8, 5.5]])
plt.gca().add_patch(plt.Polygon(X[:,:],color='blue'))
plt.imshow (main)
```

```python
import matplotlib.patches as patches
main=np.ones((12,18,3))
main[8::,::,1:]=0
main[:4:,::,0::2]=0
x = [-0.5,3,3,-0.75]
y = [-0.5,3.5,7.5,12]
plt.gca().add_patch(patches.Polygon(xy=list(zip(x,y)), color='black'))
plt.imshow (main)
```

```python
import matplotlib.patches as patches
main=np.ones((12,18,3))
circle = plt.Circle((8.5,5.5), 4, color='#dc143c')
plt.gca().add_artist(circle)
plt.imshow (main)
```

```python
import matplotlib.patches as patches
main=np.zeros((12,18,3))
x=np.array([[-0.5,-0.5], [-.5,8], [14,-0.5]])
plt.gca().add_patch(plt.Polygon(x[:,:],color='green'))
y=np.array([[18,12], [2,12], [18,3]])
plt.gca().add_patch(plt.Polygon(y[:,:],color='b'))
plt.imshow (main)
```

```python
main=np.ones((1200,1800,3))
main[600::,::,1:]=0
for i in range(600):
    for j in range(600):
        if i>=j:
            main[i:i+1,j:j+1,0:2] = 0
for i in range(600,1200,1):
            main[i:i+1,:1200-i,0:2] = 0
            main[i:i+1,:1200-i,2] = 1
plt.imshow (main)
```
