```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("seaborn-poster")
```

```python
x=np.linspace(-10,10,20)
y=x**-3
plt.plot(x,y)
plt.show()  # Not needed in inline mode
```

```python
plt.plot(x,y)
plt.xlim(0,5)
plt.ylim(0,2)
```

```python
plt.plot(x,y)
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.title("Test")
```

```python
z=4*x**-2
plt.plot(x,y,'--g')
plt.plot(x,z,':c')
plt.xlabel('x')
plt.ylabel("$f(x)$")
plt.legend(['$x^-3$','$4 x^-2$'])
```

```python
n=100
x=np.random.rand(n)
y=np.random.rand(n)
plt.scatter(x,y,s=20)
```

```python
colors=np.random.rand(n)
area=np.pi*(100*np.random.rand(n))**2
plt.scatter(x,y,s=area,c=colors,alpha=1)
```

```python
A=np.array([[[np.random.rand() for n in range(3)]for m in range(200)]for z in range(300)])
plt.imshow(A)
```
