For 2 arryas, finding common elements

```python
import numpy as np
a =np.array([1,2,3,2,3,4,3,4,5,6])
b =np.array([7,2,10,2,7,4,9,4,9,8])
c =np.intersect1d(a,b)
print(c)
```

```python
a =np.array([1,2,3,2,3,4,3,10,2,74,5,6])
b =np.array([7,2,10,2,7,4,9,4,9,8])
c =np.array([10,2,7,4,9])
from functools import reduce
reduce(np.intersect1d, (a,b,c))
```

```python
# Understand where properly before seeing this and boolean array indexing
np.random.seed(35265) # Used to maintain RandomState
A = np.random.randint(1, 100, (7, 5))
#print(A,"\n")
''' A[:,1]>50 checks elements which are greater than 50 for all rows in 2nd column, returns the result in array when we use 
    np.where '''
print(np.where(A[:,2]>50))
''' A[np.where(A[:,1]>50),1] displays the elements which are greater than 50 in 2nd column '''
print(A[np.where(A[:,2]>50),2])
print(A[np.where(A[:,4]%2==1),4])
```
