```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

np.set_printoptions(suppress=True) # for cleaner number printing
```

```python
data = pd.read_excel('dataset2 for day 1.xlsx', sheet_name='Sheet1')
x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 36)


def BuildModel(train_data, train_label, test_data, test_label, col, order = 1) :
    model = LinearRegression()
    
    train = train_data[:, col:col+1]
    test = test_data[:, col:col+1]
    
    model.fit(train ** order, train_label)
    
    print("-----------------------------------")
    print(data.columns[col+1])
    print("-----------------------------------")
    print("Order is ", order)
    print("R^2 Train ", model.score(train ** order, train_label))
    print("R^2 Test ", model.score(test ** order, test_label))

    return(model)

for val in range(data.shape[1]-2) :
    
    BuildModel(x_train, y_train, x_test, y_test, col = val, order = 1)
    BuildModel(x_train, y_train, x_test, y_test, col = val, order = 2)
    BuildModel(x_train, y_train, x_test, y_test, col = val, order = 3)
    BuildModel(x_train, y_train, x_test, y_test, col = val, order = 4)
    BuildModel(x_train, y_train, x_test, y_test, col = val, order = 5)
        
```
